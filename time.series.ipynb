```r
my.data <- read.csv ("data.csv", sep=",")
head (my.data, 10)
```

```r
cut.days.off <- function (date)
{
    tmp_store <- strsplit (date, "/")
    return (paste (tmp_store[[1]][3], tmp_store[[1]][1], sep="/"))
}

dates <- paste(my.data$DATE)
my.data$DATE <- paste(lapply (dates, cut.days.off))
head (my.data, 10)
```

```r
library ("forecast")

process.ts <- function (region.ts, unique.regions)
{
    plot (stl(log(region.ts), "per"), main=unique.regions[i])
    
    ts.fit <- auto.arima (region.ts, ic="aic")
    ts.forecast <- predict(ts.fit, n.ahead=3)  
    
    conf.up   <- ts.forecast$pred + 2*ts.forecast$se
    conf.down <- ts.forecast$pred - 2*ts.forecast$se
    max.ts=max(region.ts,conf.up)
    min.ts=min(region.ts,conf.down)
    
    ts.plot(region.ts,ts.forecast$pred, col=1:2, ylim=c(min.ts,max.ts), main=unique.regions[i])
    lines(conf.up, col="blue", lty="dashed")
    lines(conf.down, col="blue", lty="dashed") 
}

unique.dates <- unique(my.data$DATE)
unique.dates <- unlist(unique.dates[order(unlist(unique.dates))])
unique.regions <- paste(unique(my.data$BOROUGH))
unique.regions <- unique.regions[unique.regions != ""]

dates.len <- length(unique.dates)
regions.len <- length (unique.regions)

for (i in 1:regions.len)
{
    x.vec <- c(rep(NA, dates.len))
    region.df <- my.data [which(my.data$BOROUGH==unique.regions[i]), ]
    for (j in 1:dates.len)
    {
        x.vec[j] <- length(region.df[region.df$DATE==unique.dates[j], ]$DATE)
    }
    temp.start <- strsplit(unique.dates[1], "/")
    temp.start <- c(as.numeric(temp.start[[1]][1]), as.numeric(temp.start[[1]][2])) 
    temp.end <- strsplit(unique.dates[dates.len], "/")
    temp.end <- c(as.numeric(temp.end[[1]][1]), as.numeric(temp.end[[1]][2]))
    process.ts (ts(x.vec, start=temp.start, end=temp.end, frequency=12), unique.regions)
    
}
```
